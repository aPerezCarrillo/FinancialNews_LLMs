```python
import json
import os 
import pandas as pd
from sklearn.metrics import classification_report
from difflib import SequenceMatcher

# Paths to datasets
DATA_DIR = '../data/polygon'
COPY_ORIGINAL_JSON_PATH = os.path.join(DATA_DIR, 'financial_news_with_ticker_level_sentiment copy.json')
COPY_EXTRACTED_JSON_PATH = os.path.join(DATA_DIR, 'extracted_structured_data copy.json')

ORIGINAL_JSON_PATH = os.path.join(DATA_DIR, 'financial_news_with_ticker_level_sentiment.json')
EXTRACTED_JSON_PATH = os.path.join(DATA_DIR, 'extracted_structured_data.json')

# Load datasets
with open(COPY_ORIGINAL_JSON_PATH, 'r', encoding='utf-8') as f:
    original_data = json.load(f)
original_df = pd.DataFrame(original_data)

with open(COPY_EXTRACTED_JSON_PATH, 'r', encoding='utf-8') as f:
    extracted_data = json.load(f)
#extracted_df = pd.DataFrame(extracted_data)
```

```python
original_df
```

```python
# Assign unique IDs to the original dataset
original_df["id"] = original_df.index

# Ensure extracted data has the same IDs and filter out empty entries
valid_extracted_data = []
for i, entry in enumerate(extracted_data):
    if entry is None or entry.get("placeholder") == "Missing":
        print("id:" + str(i) + ", skipped")
        continue  # Skip empty entries
    entry["id"] = i  # Keep original order
    valid_extracted_data.append(entry)

extracted_df = pd.DataFrame(valid_extracted_data)
```

```python
extracted_df.columns
```

```python
extracted_df.drop(['title: ', 'title:', 're_company_details'], axis=1, inplace=True)
```

```python
extracted_df.head(20)
```

```python
# Save the updated original dataset with IDs
original_df.to_json(ORIGINAL_JSON_PATH, orient='records', indent=4)
# Save the updated extracted dataset with IDs
extracted_df.to_json(EXTRACTED_JSON_PATH, orient='records', indent=4)
print(f"✅ Processed datasets saved:\n- {ORIGINAL_JSON_PATH}\n- {EXTRACTED_JSON_PATH}")
```
